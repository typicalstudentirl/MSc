**Named Entity Recognition**: Named entities allow a further level of context. Named entities are real world objects that are assigned a name.
Spacy recognizes real world entities by asking model for a prediction. NER have an assigned type, e.g. organisation, country etc.


```python
# assign entites to array
named_entity_recs = [entity for entity in moby_dick_nlp.ents]
```

```python
# randomly chosen words: harpoon, fish, whale, wharf
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
```

```python
# match for "harpoon"
harpoon1 = [{"LOWER": "harpoon"}]
# match for "harpoons"
harpoon2 = [{"LOWER": "harpoons"}]

# match for "fish"
fish1 = [{"LOWER": "fish"}]
# match for "fishes"
fish2 = [{"LOWER": "fishes"}]
# match for "fishy"
fish3 = [{"LOWER": "fishy"}]


# match for "whale"
whale1 = [{"LOWER": "whale"}]
# match for "whalers"
whale2 = [{"LOWER": "whalers"}]
# match for "whaling"
whale3 = [{"LOWER": "whaling"}]

# match for "wharf"
wharf1 = [{"LOWER": "wharf"}]
# match for "wharf"
wharf2 = [{"LOWER": "wharves"}]

```

```python
# add matchers to one matcher object
matcher.add("Harpoon", None, harpoon1, harpoon2)

matcher.add("Fish", None, fish1, fish2, fish3)

matcher.add("Whale", None, whale1, whale2, whale3)

matcher.add("Wharf", None, wharf1, wharf2)
```

```python
# extract matches
token_matches = matcher(moby_dick_nlp)
```

```python
print(token_matches)
```

```python
# parse and print match id, word start, word end
for match_id, start, end in token_matches:
  string_id = nlp.vocab.strings[match_id]
  matched_span = moby_dick_nlp[start:end]
  print(f"{match_id:<{20}} {string_id:<{15}} {start:{3}} {end:{3}} {matched_span.text:{20}}")
```

```python
print(f"The words harpoon, whale, fish and wharf appered {len(token_matches)} times.")
```

```python
# Import the PhraseMatcher library
from spacy.matcher import PhraseMatcher
phrase_matcher = PhraseMatcher(nlp.vocab)
```

```python
# create a new phrase list using words from q7
phrase_list = ["harpoon", "whale", "fish", "wharf"]
```

```python
# create phrase patterns array
phrase_patterns = [nlp.make_doc(word) for word in phrase_list]
print(phrase_patterns)
```

```python
phrase_matcher.add("NLP", None, *phrase_patterns)
matches = phrase_matcher(moby_dick_nlp)
print(matches)
```

```python
# parse and print match id, word start, word end
for match_id, start, end in matches:
  string_id = nlp.vocab.strings[match_id]
  span = moby_dick_nlp[start:end]
  print(match_id, "\t", string_id, "\t", start, "\t", end, "\t", span.text)
```

```python
print(f"The words harpoon, whale, fish and wharf appered {len(matches)} times.")
```
